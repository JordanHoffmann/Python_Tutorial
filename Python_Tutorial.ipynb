#LIST/ARRAYS

When dealing with data, you often want to keep track of multiple numbers at once, either in the form of a matrix or an array. There are many ways to do this in Python. Perhaps the most straightforward is the very flexible list construction that Python has built in.

```python
a = [1,2,3]
print a
print a[0]
print 'Min is: ', min(a)
print 'Max is: ', max(a)
```

However, the list construction does have multiple limitations that you might take for granted coming from MATLAB or Mathematica. For example:

```python
a**2
```

One way around this is:

```python
a_sq = [ a[i]**2 for i in range(len(a)) ]
print a_sq
```

```python
a = np.array([1,2,3])
print a
print a[0]
```

```python
print a**2
```

The question of is one better than the other is hard to answer. You might be interested in many different things.
+ Speed
+ Portabiltiy of the Code
+ Readability
By using numpy, for example, to run your code someone else must also have numpy. We can do some timing tests:

```python
import time as time
number_of_trials = int(1e7)
test_array = range(number_of_trials)
t0 = time.time()
test_sq_1 = [ test_array[i]**2 for i in range(number_of_trials) ]
t1 = time.time()
delta_t = t1-t0
test_nparray = np.array(range(number_of_trials))
t2 = time.time()
test_sq_2 = test_nparray**2
t3 = time.time()
delta_t2 = t3-t2
speedup = delta_t/delta_t2
print 'The list method took: %f seconds' % delta_t
print 'The numpy array method took: %f seconds' % delta_t2
print 'The numpy method was %f times faster!!!' % speedup
```

```python
def polynomial_1(x):
    return x**3+x**2+x

def polynomial_2(x,y):
    return x*y-x**1.3-y

def func_3(n_max):
    yes = 0
    no = 0
    for i in xrange(n_max):
        r = np.random.rand()
        if r < 0.5:
            yes = yes + 1
        else: 
            no = no+1
    return float(yes)/n_max
```

###Can call a function recursively

```python
def factorial(n):
    if n<1: 
        return 1
    else:
        tmp=n*factorial(n-1)
        return tmp
```

#While Loops via Gradient Descent

One common theme of this course is that we would like to continue a procedure until the error is below some prescribed value. To give a simple example of how to do this, let's consider a simple greedy, gradient descent algorithm.

First, we define a simple function that we would like to find the minimum of.

```python
def f(x):
    return x**2+(x)-5*x-3
def df(x):
    return 2*x+1-5
```

We setup 

```python
xs=np.linspace(-4,10,100)
```

```python
ys=f(xs)
```

```python
plt.plot(xs,ys,'r-')
```

```python
#setup a tolerance we will accept
error = 1e-6
```

```python
start = 0
current = 14*np.random.rand()-4
error = 1e-6
alpha = 0.1
iteration = 0

while abs(current - start) > error:
    start = current
    current = start - alpha * df(start)
    iteration+=1

print "Local minimum occurs at", current
print "It took %d iterations" %iteration
```

#Example: Pi a la Monte Carlo

```python
#Define variables
N = 10000 #number of samples
```

```python
xs = np.random.rand(N)
ys = np.random.rand(N)

#Using append is not very fast, but it is very readable
#and, in this case, it is fast enough

inside = []
outside = []

#We want to check if the value is inside or outside
for i in range(N):
    if xs[i]**2+ys[i]**2<=1:
        inside.append([xs[i],ys[i]])
    else:
        outside.append([xs[i],ys[i]])
```

```python
[x1,x2] = np.array(inside).T
[x3,x4] = np.array(outside).T
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.set_aspect('equal')
ax.plot(x1,x2,'ro',label='Inside')
ax.plot(x3,x4,'bo',label='Outsde')
legend = plt.legend()
frame = legend.get_frame()
frame.set_facecolor('1')
#plt.legend(loc='best')
plt.show()
```

```python
def mc_pi(N):
    xs = np.random.rand(N)
    ys = np.random.rand(N)

    #Using append is not very fast, but it is very readable
    #and in this case, it is fast enough

    inside = []
    outside = []

    #We want to check if the value is inside or outside
    for i in range(N):
        if xs[i]**2+ys[i]**2<=1:
            inside.append([xs[i],ys[i]])
        else:
            outside.append([xs[i],ys[i]])
    return 4*len(inside)/float(N)
```

```python
log_space = np.logspace(6,2).astype(int)
print log_space
```

```python
pi_vals = np.zeros(len(log_space))
```

```python
for i in xrange(len(log_space)):
    pi_vals[i] = mc_pi(log_space[i])
```

```python
plt.plot(log_space,abs((pi_vals-np.pi))/np.pi)
```

```python
plt.loglog(log_space,abs((pi_vals-np.pi))/np.pi)
```

#Tennis

```python
player1=0.0
player2=0.0
```

```python
def game(p):
    p1=0
    p2=0
    flag=0
    while flag==0:
        ran=np.random.rand()
        if ran < p:
            p1+=1
        else:
            p2+=1
        if (p1>=4 or p2>=4) and abs(p1-p2)>=2:
            flag=1
    return p1-p2
```

```python
number_of_samples = 1000
data=[[o/float(number_of_samples), np.mean(np.array([game(o/float(number_of_samples)) for x in xrange(number_of_samples)]))] for o in xrange(number_of_samples)]
```

```python
xs=np.transpose(data)[0]
ys=np.transpose(data)[1]

plt.plot(xs,ys)
plt.title('$\mathrm{ Single \; game \; to  \; 4 \; win \; by \; 2 }$')
plt.xlabel('$\mathrm{Probability \; that \; player \; wins \; a \; point}$')
plt.ylabel('$\mathrm{Expected \; difference \; in  \; score \; of \; game }$')
```

```python
def set(p):
    player1=0
    player2=0
    flag=0;
    while flag==0:
        gam=game(p)
        if gam > 0:
            player1+=1
        else:
            player2+=1
        if (player1>=6 or player2>=6) and abs(player1-player2)>=2:
            flag=1
    return player1-player2
```

```python
number_of_samples = 1000
dataset=[[o/float(number_of_samples), np.mean(np.array([set(o/float(number_of_samples)) for x in xrange(number_of_samples)]))] for o in xrange(number_of_samples)]
```

```python
xs2=np.transpose(dataset)[0]
ys2=np.transpose(dataset)[1]
plt.plot(xs2,ys2,'k-')
plt.title('$\mathrm{ Set \; game \; to  \; 6 \; win \; by \; 2 }$')
plt.xlabel('$\mathrm{Probability \; that \; player \; wins \; a \; point}$')
plt.ylabel('$\mathrm{Expected \; difference \; in  \; score \; of \; set }$')
```
